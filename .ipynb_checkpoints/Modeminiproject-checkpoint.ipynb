```python
from sqlalchemy import create_engine, text
from sqlalchemy.engine import URL

url = URL.create(
    "postgresql+psycopg2",
    username="postgres",
    password="Abhinesh4533@",   # replace with your actual password
    host="127.0.0.1",
    port=5432,
    database="modesales",             # connect to default DB first
)
engine = create_engine(url)

with engine.connect() as conn:
    result = conn.execute(text("SELECT version();"))
    print(result.fetchone())

import pandas as pd
from sqlalchemy import types as satypes

df = pd.read_csv(r"C:/Users/abhin/Downloads/Job Search/Python/Mode/sales_data.csv", parse_dates=["order_date"])

dtype_map = {
    "customer_id":  satypes.Integer(),
    "order_id":     satypes.Integer(),
    "order_date":   satypes.Date(),
    "product":      satypes.Text(),
    "sales_amount": satypes.Numeric(12, 2),
}

df.to_sql("sales", engine, if_exists="replace", index=False, dtype=dtype_map)
print("Rows loaded:", len(df))

```

```python
# (Top 5 customers)

top_customers_sql = """
SELECT customer_id, ROUND(SUM(sales_amount), 2) AS total_sales
FROM sales
GROUP BY customer_id
ORDER BY total_sales DESC
LIMIT 5;
"""
top_customers = pd.read_sql(top_customers_sql, engine)
top_customers
```

```python
# Python analysis (monthly trend)
df_sql = pd.read_sql("SELECT * FROM sales;", engine, parse_dates=["order_date"])
df_sql["order_month"] = df_sql["order_date"].dt.to_period("M").dt.to_timestamp()

monthly_sales = (
    df_sql.groupby("order_month", as_index=False)["sales_amount"]
          .sum()
          .rename(columns={"sales_amount":"monthly_sales"})
)
monthly_sales
```

```python
# Charts (Matplotlib)
import matplotlib.pyplot as plt

# Monthly trend
plt.figure()
plt.plot(monthly_sales["order_month"], monthly_sales["monthly_sales"], marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month"); plt.ylabel("Total Sales")
plt.xticks(rotation=45); plt.tight_layout()
plt.savefig("monthly_sales_trend.png", dpi=150)
plt.show()
```

```python
# Top customers bar chart
plt.figure()
plt.bar(top_customers["customer_id"].astype(str), top_customers["total_sales"])
plt.title("Top 5 Customers by Total Sales")
plt.xlabel("Customer ID"); plt.ylabel("Total Sales")
plt.tight_layout()
plt.savefig("top_customers.png", dpi=150)
plt.show()
```
